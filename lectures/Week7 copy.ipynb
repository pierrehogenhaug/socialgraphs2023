**Step two: Word weights**

TF is not necessarily a good way of sorting a list, since many words are very common, so the most common words are not necessarily the most important ones. This is clear from the top counts in my west coast TF list

```
the 17309
and 9778

 ```
 
You can fix some of this by removing stop-words (as is described in the book), but it's not enough. ***We want to pull out what's important.*** Thus, to create awesome and informative wordclouds like the ones I created above, we want to extract what's special about each of the two lists we're comparing. 

The general way of doing that is using a strategy called TF-IDF. We will explore that strategy in exercise 4 below. But out-of-the-box TF-IDF gets a bit weird when we only have two groups to compare. So we'll do something different (but related) here. 

We want to use the information stored in TF, which is important since it says something about the most frequently occuring words. **But we want to weigh the TF information** using additional information about what is unique about each coast. 

Specifically, we want to set things up such that - in the west-coast list, for example - words that are unique to the West Coast get a high weight, while words that are unique to the East Coast get a low weight, and words that occur at the same frequency in both lists are neutral. Vice versa for the west-coast list.

The way I chose to set up the weights here is simple. I use term ratios (TR), which I just made up, so don't look it up on the internet. What we care about is words that are used very differently for the two coasts, so we just use their ratios. 

**normalization..** Let me start with an example. The word `california` occurs 458 in the West-coast TF list and 87 times in the East-coast TF list, thus I set it's West-coast weight to

$$w_\textrm{california}^{(m)} = \frac{458}{87 + c} = 4.28.$$

Similarly, its East-coast weight is 

$$w_\textrm{california}^{(d)} = \frac{87}{458 + c} = 0.18.$$

In both cases, I add the constant $c$ to the denominator in case a word occurs zero times. You can play around with the size of $c$ to understand the effect of chosing small/large values of it.

In general for some token $t$ with term frequency $\textrm{TF}^{(u)}_t$, in coast $u$ where $u \in \{w,e\}$, we define the weight as:

$$w_{t}^{(w)} = \frac{\textrm{TF}^{(w)}_t}{ \textrm{TF}^{(e)}_t + c}, \qquad \textrm{and} \qquad w_{t}^{(e)} = \frac{\textrm{TF}^{(e)}_t}{ \textrm{TF}^{(w)}_t + c}. $$

Thus, now we're ready to lists for each univers, where the ranking of token $t$ on list $u$ is given by $\textrm{TF}^{(u)}_t \times w_{t}^{(u)}$. I call this the *TF-TR* lists. 

Note that the *TF-TR* lists have the properties that we requested above. The weight associated with a word is large when a word occurs much more frequently in the list we consider, compared to the other list. It's small when a word is rare in our list and frequent in the other. And it's approximately equal to one, when the word is equally frequent in both lists. (The downside is that it only works when you're finding important terms while comparing two lists).

The takehome here is that we get a good ranking by combining the term frequency with some weighting scheme. But, as we will see below, the TR weight that I created is not the only possible weight. There are many other options.

-----------

PS. Above in creating the ratios, I'm assuming that the coast pages have a comparable number of words in them. That's pretty much true (up to a factor of 2). And while the approximation is OK for our purposes, it's not true in general. If you'd like an extra challenge, you can figure out how to account for differences in the size of each coast corpus.

````python
# Convert the TF lists to dictionaries
tf_dict_east_coast = dict(tf_list_east_coast)
tf_dict_west_coast = dict(tf_list_west_coast)

# Constant c
c = 1 

def compute_weights(tf_dict_a, tf_dict_b, c):
    weights = {}
    for token, tf in tf_dict_a.items():
        weights[token] = tf / (tf_dict_b.get(token, 0) + c)
    return weights

weights_east_coast = compute_weights(tf_dict_east_coast, tf_dict_west_coast, c)
weights_west_coast = compute_weights(tf_dict_west_coast, tf_dict_east_coast, c)
````

````python
# Compute TF-TR for each coast
def compute_tf_tr(tf_dict, weights):
    tf_tr = {}
    for token, tf in tf_dict.items():
        tf_tr[token] = tf * weights[token]
    return tf_tr

tf_tr_east_coast = compute_tf_tr(tf_dict_east_coast, weights_east_coast)
tf_tr_west_coast = compute_tf_tr(tf_dict_west_coast, weights_west_coast)

# Sort and get the top 10 for each coast
sorted_tf_tr_east_coast = sorted(tf_tr_east_coast.items(), key=lambda x: x[1], reverse=True)
sorted_tf_tr_west_coast = sorted(tf_tr_west_coast.items(), key=lambda x: x[1], reverse=True)

print("Top 10 for East Coast:")
for token, value in sorted_tf_tr_east_coast[:10]:
    print(token, value)

print("\nTop 10 for West Coast:")
for token, value in sorted_tf_tr_west_coast[:10]:
    print(token, value)
````

````python
# convert sorted TF-TR lists into a single string by repeating each word according to its score

def tf_tr_to_string(sorted_tf_tr):
    word_list = []
    for word, score in sorted_tf_tr:
        word_list.extend([word] * round(score))
    return ' '.join(word_list)

string_east_coast = tf_tr_to_string(sorted_tf_tr_east_coast)
string_west_coast = tf_tr_to_string(sorted_tf_tr_west_coast)


# Add any additional stopwords if needed
stopwords = set(STOPWORDS)
````

````python
### Code from Week 4

import networkx as nx
import pandas as pd
import pickle as pkl
import regex as re


with open('rapper_dict.pkl', 'rb') as fp:
    all_rapper_pages = pickle.load(fp)


# URLs of the CSV files
west_coast_url = 'https://raw.githubusercontent.com/SocialComplexityLab/socialgraphs2023/main/files/WestCoastRappers.csv'
east_coast_url = 'https://raw.githubusercontent.com/SocialComplexityLab/socialgraphs2023/main/files/EastCoastRappers.csv'

# Reading the CSV files into pandas DataFrames
west_coast_df = pd.read_csv(west_coast_url)
east_coast_df = pd.read_csv(east_coast_url)

west_coast_rappers = []
east_coast_rappers = []

# Loop through the west coast rapper URLs and print the content
for _, row in west_coast_df.iterrows():
    rapper_name = row['WikipediaPageName']
    west_coast_rappers.append(rapper_name)

# Repeat for the east coast rappers
for _, row in east_coast_df.iterrows():
    rapper_name = row['WikipediaPageName']
    east_coast_rappers.append(rapper_name)

# All rappers list - combine both West and East coast rapper names
all_rappers = west_coast_rappers + east_coast_rappers

# Build the Network
rapper_network = nx.DiGraph()

# Adding nodes to the graph with coast attribute
for rapper_name in west_coast_rappers:
    rapper_network.add_node(rapper_name, coast='West')

for rapper_name in east_coast_rappers:
    rapper_network.add_node(rapper_name, coast='East')
    
# Extract Outgoing Links Using Regular Expression
link_re = re.compile(r'\[\[(.*?)(\|.*?)*\]\]')

# Function to extract links and build the network
def process_page(rapper_name, page_content):
    links = link_re.findall(page_content)
    for link in links:
        # Extracting the linked rapper's name
        linked_rapper = link[0].replace(' ', '_')
        if linked_rapper in all_rappers:
            rapper_network.add_edge(rapper_name, linked_rapper)

# Iterating through each rapper's page and processing it
for rapper_name, page_content in all_rapper_pages.items():  # You need to load all rapper pages into this dictionary
    process_page(rapper_name, page_content)

# Removing nodes with no in- and out-degrees
isolated_nodes = [node for node, degree in rapper_network.degree() if degree == 0]
rapper_network.remove_nodes_from(isolated_nodes)

# Filtering by coast
east_coast_nodes = [n for n, d in rapper_network.nodes(data=True) if d['coast'] == 'East']
west_coast_nodes = [n for n, d in rapper_network.nodes(data=True) if d['coast'] == 'West']
````

Steps to calculate **Modularity**

- Compute L: Total number of links (edges) in the entire network.
- Compute Lc: Number of links between nodes within each community.
- Compute kc: Sum of the degrees of nodes in each community.

````python
# Step 1
L = rapper_network.number_of_edges()

# Step 2 + 3 
Lc_east = sum(1 for u, v in rapper_network.edges() if u in east_coast_nodes and v in east_coast_nodes)
kc_east = sum(rapper_network.degree(n) for n in east_coast_nodes)

Lc_west = sum(1 for u, v in rapper_network.edges() if u in west_coast_nodes and v in west_coast_nodes)
kc_west = sum(rapper_network.degree(n) for n in west_coast_nodes)

# Calculate M
M = (Lc_east / L - (kc_east / (2 * L)) ** 2) + (Lc_west / L - (kc_west / (2 * L)) ** 2)
M
````

If M is close to zero, the division isn't particularly meaningful, and there's not a clear community structure based on the coasts.
